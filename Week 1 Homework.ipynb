# Discrepancies Between Describe and Shape
When using the shape attribute of the DataFrame, we see that it has 891 rows and 15 columns but when using the describe method, we see that some of the columns have less than 891 entries. This is because the describe method does not count the non-numeric or the missing datapoints.

# Difference Between Attributes and Methods
- Attributes are properties that belong to an object. shape is an attribute because it belongs to the DataFrame object called df2.
- Methods are functions that are associated with the object. describe() is a method because it calls a functions belonging to the DataFrame object which then calculates things sich as the mean, standard deviation, max/min, etc.

# DataFrame.describe() Variable Meanings
- Count: The number of non-null entries in this column.
- Mean: The sum of the column divided by the number of entries (the average entry).
- std: The standard deviation. A way to measure how far the datapoints lie from their trendline.
- min: The smallest value in the column.
- 25%: The first quartile. 25% of the data falls below this value.
- 50%: The median (aka the second quartile). 50% of data falls below this value.
- 75%: The third quartile. 75% of data falls below this value.
- max: The largest value in the column.

# `df.dropna()` vs `del df['col']`
df.dropna() will drop all rows that contain missing data from the dataframe. whereas del df['col'] will completely remove that column from all rows.


1. **Provide an example of a "use case" in which using df.dropna() might be peferred over using del df['col'].**

* For our initial dataset with the Animal Crossing NPCs, if we want to perform some analysis on the NPCs songs, we can use `df.dropna()`. Some NPCs in the dataframe do not have an entry for 'song'. We can use dropna() to remove those NPCs from our database and still continue with our analysis of the remaining NPCs songs. Using del df['col'] would remove all the NPCs songs from our dataframe, making it impossible for us to conduct the analysis.

2. **Provide an example of "the opposite use case" in which using del df['col'] might be preferred over using df.dropna()**

* In our Animal Crossing NPC dataset, there is a 'url' column which contains a url to an image of the character. Suppose we do not need to perform any alaysis on the appearance of the character. In this senario, we can use `del df['url']` to remove this column from our dataframe. This could be useful if we had an extremely large dataset that is taking up an unessesary amount of space.

3. **Discuss why applying del df['col'] before df.dropna() when both are used together could be important**

* When using both del df['col'] and df.dropna(), we must apply del df['col'] first because if a row had missing data in the column that we are deleting, the row is still useful because we do not care about what is in that column. If an NPC does not have an entry in their 'url' column but they have everything, we can still use that npc to perform our analysis on the songs. If we use dropna(), it will remove them from the dataset but if we delete the column first, then there will not be any null datapoints for dropna() to detect.
